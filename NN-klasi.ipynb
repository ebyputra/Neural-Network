```python
import pandas as pd
import numpy as np

df = pd.read_excel('klasi.xlsx')
df
```

```python
df.info()
```

```python
N, D = df.shape
print('N: %d, D: %d' %(N, D))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Pilek'] = encoder.fit_transform(df['Pilek'])
df['Demam'] = encoder.fit_transform(df['Demam'])
df['Batuk'] = encoder.fit_transform(df['Batuk'])
df['Sesak Nafas'] = encoder.fit_transform(df['Sesak Nafas'])
df['Keluhan Lain'] = encoder.fit_transform(df['Keluhan Lain'])
```

```python
df
```

```python
def ubah(v):
    if v == 1:
        return 0
    else:
        return 1
df['Pilek'] = df.Pilek.apply(ubah)

def ganti(w):
    if w == 1:
        return 0
    else:
        return 1
df['Gender'] = df.Gender.apply(ganti)
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df, test_size=0.25, random_state=2)
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

```python
print(X_train)
```

```python
X_train1=X_train.values
X_train1=np.delete(X_train1, 9,axis=1)
y_train1=X_train['Status'].values
print(X_train1.shape)
print(X_train1)
```

```python
X_test1=X_test.values
X_test1=np.delete(X_test1, 9, axis=1)
y_test1= X_test['Status'].values
print(X_test1.shape)
print(y_test1)
```

```python
from sklearn.preprocessing import MinMaxScaler
X_train1=MinMaxScaler().fit_transform(X_train1)
X_test1=MinMaxScaler().fit_transform(X_test1)
print(X_train1)
print(X_test1)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(9, activation='relu', input_shape=(9,)))

model.add(Dense(10, activation='relu'))


model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train1, y_train1, epochs=200, verbose=1, batch_size=20)

y_pred = model.predict(X_test1)
y_pred = y_pred >= 0.15
```

```python
y_test
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(hidden_layer_sizes=10, learning_rate_init=0.1, max_iter=100)
clf.fit(X_train1, y_train1)
```

```python
y_pred=clf.predict(X_test1)
from sklearn.metrics import accuracy_score,confusion_matrix
round(accuracy_score(y_test1, y_pred), 3)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test1, y_pred))
```

```python
confusion_matrix(y_test1, y_pred)
```
